## Training dataset generation

In this stage the training dataset is generated using Satproc. This tool generates the dataset with characteristics needed for the model: size, number of images, value range, etc.

### Satproc

Satproc_extract_chips tool takes the image and the vector file with the annotations and generates a directory with two folders: "images" and "extent". Files in the first one are Tiff files with 10-bands. Also, files from the second folder are Tiff type but with 1-band, because the number of classes . Each mask is a binary image.

Observation: images and the vector file (gt) should have the same georeference, for instance, EPSG:4326

```python
path_to_files= f'{PATH_TRAIN_IMAGES}*.tif' # images folder
output_folder=os.path.join('./../dataset/data_train/',ZONE,RESO,str(SIZE)+'_'+str(STEP_SIZE)+'/')  #dataset destination folder
vector_file = f'{PATH_GT}{GT_FILE}' #  ground truth vector file 
vector_file_aoi = vector_file#f'{PATH_GT}{AOI_FILE}' #area of interest vector file
```

```python
print(path_to_files,output_folder,vector_file,vector_file_aoi)
```

Satproc execution

```python
#!satproc_extract_chips --help
```

```python
!satproc_extract_chips \
                $path_to_files \
                -o $output_folder \
                --size $SIZE \
                --step-size $STEP_SIZE \
                --aoi $vector_file_aoi \
                --labels $vector_file \
                --label-property 'class1' \
                --classes 'D' \
                --rescale \
                --rescale-mode $RESCALE_MODE
```

```python
ls ./../dataset/data_train/TEGUCIGALPA/low/160_40/images/ | wc -l
```

##### **Arguments:**

* **path_to_files:** images path

* **o:** destination path



We recommend use a descriptive path. For instances: "./../data_train/120_40/" describes: data_train  data use for train; 120_40  < image size >_< step size >.

* **size:** chips size (chips are square images)
* **step-size:** process step. It should be less or equal to 'size'. If it's equal to size, there isn't an overlap in the resulting images 

 *Sometimes, it's useful generating overlaping images for the training, because the overlaping increase the number of images in the training dataset*.


* **label-property:** field name where the classes are defined (it's only used for training dataset).

* **classes:** classes names (as they appear in the geojson file), separated by spaces

* **aoi:** aoi file path. When an area of interest(area where annotations are defined) is stablished, Satproc only process that area of the image

* **rescale:** rescale band values to 0-255

This command generate two folder in the destination path: "images" and "extent". Files in the first one are Tiff files with 3-bands (RGB). Secondly, files from the second folder are Tiff type but with N-bands, where N is the number of classes. Each band is a binary image.

## Training

We train the model using the datasets (images and masks) created with Satproc from which the model learn about the patterns that defines the object of interest.

In this stage, we define the model configuration.

Obs: It's useful named the model using information about training parameters. For instance: < model >< proyect >< images_dimension >< size >< step_size >_< step_per_epoch >.h5 or similar.

```python
config = TrainConfig(width   = INPUT_SIZE,  #  tamao de la imagen procesada por la UNet (debe ser multiplos de 16 , por ej 160, 320,etc; y no menor a 80) // image size processed by the Unet (It must be a multiple of 16 bigger than 80)
                     height  = INPUT_SIZE,
                     n_channels   = N_CHANNELS,  #  nmero de canales de la imagen, rgb -> 3 // number of bands of the image
                     n_classes    = N_CLASSES, # nmero de clases a clasificar// number of classes to classify
                     apply_image_augmentation = True, #  si es True, amplia el dataset generando imagenes nuevas a partir de pequeas variaciones de las ya existentes (rotacin,)//If it's True, increase the dataset size by the creation of new images from little modification of the orginal images (rotation) 
                     seed       = 42,
                     epochs     = 20, # Cantidad de veces que el dataset entero puede pasar por el proceso de entrenamiento// Number of times the full dataset could be processed in the training
                     batch_size = 16, #Cantidad de datos que se procesan por vez, puede ser limitado por la memoria de gpu disponible (debe ser multiplo de 16)//  Number of training examples utilized in one iteration.
                     steps_per_epoch  = STEP_PER_EPOCH, #  Tpicamente debe ser igual al numero de imgenes / el batch_size, si es mayor incrementara el nmero de imgenes generadas con image augmentation. // Normally, is equal to the ratio between the dataset size and the batch size. If it's bigger, increase the number of images created by image_augmentation .
                     early_stopping_patience  = 6, # a medida que entrena se guardan los resultados del entrenamiento despues de cada epoch, si el error no vari luego de 5 iteraciones , se corta el proceso porque se entiende que el error ya disminuy significativamente. // After each epoch, the training results are saved. If there isn't variation of the error 5 iterations after the saved model, the training proccess is over.
                     validation_split   = 0.1, # se divide la muestra en training (90% de los datos) y validation (10% de los datos) para calcular el error durante el proceso de entrenamiento.// The training dataset is divided: 90% to train the model and 10% to validate it (to calculate the error).
                     test_split         = 0.1, # Cantidad de imgenes del dataset test // Number of images in the test dataset
                     images_path        = output_folder,#ruta a las imgenes generadas con Satproc //Satproc images destination path
                     model_path         = os.path.join(BASE_PATH,'data/weights/', model_name),# ruta del modelo entrenado // Trained model path
                     model_architecture = 'unet', #Arquitectura de la red // Neural network architecture
                     evaluate           = True  ,
                     class_weights      = CLASS_WEIGHT) 
```

We can visualize some images and mask from the training dataset. (Left: image, Right: mask.)

```python
plot_data_generator(num_samples=3, fig_size=(20, 20), train_config=config,img_ch =3)
```

Running the training process

```python
res_config = train(config)
```

Plotting the metrics

```python
plt.figure(figsize=(16,4))

plt.subplot(121)
plt.plot(res_config.history['loss'])
plt.plot(res_config.history['val_loss'])
plt.title('Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper left')

plt.subplot(122)
plt.plot(res_config.history['mean_io_u'])
plt.plot(res_config.history['val_mean_io_u'])
plt.title('mean_iou')
plt.ylabel('val_mean_iou_1')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper left')


plt.show()
```
